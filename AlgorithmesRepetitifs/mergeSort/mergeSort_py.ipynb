### NaiveMethod

```python
def mergeSort(A): # main ; but I use recursive so I name "merge" to more comprehension
    if len(A) > 1: # while we can divide A into 2 parts
        m = len(A)//2
        L = A[:m]
        R = A[m:]

        mergeSort(L)
        mergeSort(R)
 
        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] <= R[j]:
                A[k] = L[i]
                i += 1
            else:
                A[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            A[k] = L[i]
            i += 1
            k += 1
 
        while j < len(R):
            A[k] = R[j]
            j += 1
            k += 1
 
        return A
```

```python

def merge(A, Aux, l, r):
    if (r > l):
        m = (l+r) // 2
        # Divise à gauche et à droite
        merge(Aux, A, l, m) 
        merge(Aux, A, m + 1, r)

        i = l
        j = m + 1
        for k in range(l,r+1):
            # Fusion (merge) des parties 
            if i > m:
                Aux[k] = A[j]
                j += 1
            elif j > r:
                Aux[k] = A[i]
                i += 1
            # Tri
            elif A[j] < A[i]:
                Aux[k] = A[j]
                j += 1
            else:
                Aux[k] = A[i]
                i += 1



def mergeSort(A):
    Aux = A.copy()
    merge(Aux, A, 0, len(A)-1)
```

```python
%%cython
import cython
import numpy as np
cimport numpy as cnp

# nogil permet de gagner quelques % de performance (à revérifier) 
@cython.boundscheck(False)
@cython.wraparound(False)
@cython.initializedcheck(False)
@cython.binding(False)
cdef void mergeSort(cnp.int64_t [:] A, cnp.int64_t [:] Aux, Py_ssize_t l, Py_ssize_t r) nogil:
    cdef Py_ssize_t i, j, k, m
    if (r > l):
        m = (l+r)//2
        mergeSort(Aux, A, l, m) 
        mergeSort(Aux, A, m + 1, r)   

        i = l
        j = m + 1
        for k in range(l, r + 1):
            # Fusion (merge) des parties 
            if i > m:
                Aux[k] = A[j]
                j += 1
            elif j > r:
                Aux[k] = A[i]
                i += 1
            # Tri
            elif A[j] < A[i]:
                Aux[k] = A[j]
                j += 1
            else:
                Aux[k] = A[i]
                i += 1
        
cpdef merge_sort_main_cython(A):
    Aux = np.copy(A)
    mergeSort(Aux, A, 0, len(A) - 1)
```
