```python
import numpy as np

def main(x, y, alpha, n):
    theta0 = theta1 = 0
    Grad0 = Grad1 = 0
    m = len(x)
    am = alpha/m
    for i in range(n):
        for j in range(m):
            ttxy = theta0 + theta1 * x[j] - y[j]
            Grad0 += ttxy 
            Grad1 += ttxy * x[j]
        theta0 -= am*Grad0
        theta1 -= am*Grad1
        Grad0 = Grad1 = 0

    return theta0, theta1
```
